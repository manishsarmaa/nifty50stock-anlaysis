```python
df1=pd.read_csv('merged_file.csv')
```

```python
df1.head()
```

__SMA__

```python
df1['MA50'] = df1['Close'].rolling(window=50).mean()
df1['MA200'] = df1['Close'].rolling(window=200).mean()
```

```python
df1.head()
```

__Moving Avg__

```python
df1['EMA 12'] = df1['Close'].ewm(span=12, adjust=False).mean()
df1['EMA 26'] = df1['Close'].ewm(span=26, adjust=False).mean()
```

```python
df1.head()
```

 __MACD - Moving Average Convergence Divergence__

```python
df1['MACD'] = df1['EMA 12'] - df1['EMA 26']
```

```python
df1['Signal_Line'] = df1['MACD'].ewm(span=9, adjust=False).mean()
```

```python
df1['Histogram'] = df1['MACD'] - df1['Signal_Line']
```

```python
df1.head()
```

__On Balance Volume (OBV)__

```python
df1['Daily Returns'] = (df1['Close'] - df1['Prev Close']) / df1['Prev Close'] * 100
```

```python
df1['Direction'] = [1 if x >= 0 else -1 for x in df1['Daily Returns']]
```

```python
df1.head()
```

```python
df1['OBV'] = (df1['Volume'] * df1['Direction']).cumsum()
```

__Bollinger Bands__

```python
rolling_mean = df1['Close'].rolling(window=20).mean()  
rolling_std = df1['Close'].rolling(window=20).std()    
df1['Bollinger Upper Band'] = rolling_mean + (rolling_std * 2)  
df1['Bollinger Lower Band'] = rolling_mean - (rolling_std * 2) 
```

__ADX(Avg. Directional Index)__

```python
df1['TR'] = np.maximum(
    df1['High'] - df1['Low'],
    np.maximum(
        abs(df1['High'] - df1['Prev Close']),
        abs(df1['Low'] - df1['Prev Close']) 
    )
)
```

```python
df1['+DM'] = np.where(
    (df1['High'] - df1['High'].shift(1)) > (df1['Low'].shift(1) - df1['Low']),
    np.maximum(df1['High'] - df1['High'].shift(1), 0),
    0
)
df1['-DM'] = np.where(
    (df1['Low'].shift(1) - df1['Low']) > (df1['High'] - df1['High'].shift(1)),np.maximum(df1['Low'].shift(1) - df1['Low'], 0),0)
```

```python
df1['TR smoothed'] = df1['TR'].rolling(window=14).sum()
df1['+DM smoothed'] = df1['+DM'].rolling(window=14).sum()
df1['-DM smoothed'] = df1['-DM'].rolling(window=14).sum()
```

```python
df1['+DI'] = (df1['+DM smoothed'] / df1['TR smoothed']) * 100
df1['-DI'] = (df1['-DM smoothed'] / df1['TR smoothed']) * 100
```

```python
df1['DX'] = (abs(df1['+DI'] - df1['-DI']) / (df1['+DI'] + df1['-DI'])) * 100
df1['ADX'] = df1['DX'].rolling(window=14).mean()
```
